```python
import glob
```

```python
def display_images(data, no_imgs_masks):
    images = glob.glob('/media/Ubuntu/meg/mlprojects/Semantic_Segmentation_for_self_driving_cars/archive/'+data+'/'+data+'/CameraRGB/*.png')[:no_imgs_masks]
    masks = glob.glob('/media/Ubuntu/meg/mlprojects/Semantic_Segmentation_for_self_driving_cars/archive/'+data+'/'+data+'/CameraSeg/*.png')[:no_imgs_masks]


    for i in range(len(images)):
        images[i] = cv2.imread(images[i])
        masks[i] = cv2.imread(masks[i], cv2.IMREAD_GRAYSCALE)


    fig, axes= plt.subplots(no_imgs_masks,2)

    for i in range(len(masks)):
        axes[i][0].imshow(images[i])
        axes[i][1].imshow(masks[i])
    plt.show()
```

```python
display_images('dataA', 3)
```
